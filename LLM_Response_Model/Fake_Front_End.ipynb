```python
import json, jwt, websocket, threading, time, datetime as dt
import sys

# ---------- 1. å…ˆç­¾ä¸€ä¸ªåˆæ³• JWT ----------
SECRET     = "kunar"          # admin-secret-key
ALGORITHM  = "HS256"
STUDENT_ID = "1001"
CHAT_ID    = "2001"

def gen_token(sub: str, hours: int = 2) -> str:
    now = dt.datetime.utcnow()
    payload = {"sub": sub, "iat": now, "exp": now + dt.timedelta(hours=hours)}
    return jwt.encode(payload, SECRET, algorithm=ALGORITHM)

TOKEN = gen_token(STUDENT_ID)
print("JWT =", TOKEN)

# ---------- 2. å»ºç«‹ WebSocket --------------
WS_URL = f"wss://02fc-120-237-20-203.ngrok-free.app/websocket/{STUDENT_ID}/{CHAT_ID}"

def on_open(ws):
    print("\nâœ… å·²è¿æ¥ï¼šè¾“å…¥å†…å®¹å›è½¦å‘é€ï¼Œexit é€€å‡º\n")

def on_msg(ws, message):
    # æ¯æ¬¡æ”¶åˆ°ä¸€æ®µæ¶ˆæ¯å°±è¿½åŠ åˆ°å½“å‰è¡Œ
    print(message, end="", flush=True)

def on_error(ws, err):
    print("\nâŒ é”™è¯¯ï¼š", err)

def on_close(ws, code, reason):
    print(f"\nğŸ”’ è¿æ¥å…³é—­ code={code}, reason={reason}")

ws_app = websocket.WebSocketApp(
    WS_URL,
    subprotocols=[TOKEN],  # â† JWT æ”¾è¿™é‡Œ
    on_open=on_open,
    on_message=on_msg,
    on_error=on_error,
    on_close=on_close,
)

# è®© WebSocket åœ¨åå°çº¿ç¨‹è·‘
def run_ws():
    ws_app.run_forever()

threading.Thread(target=run_ws, daemon=True).start()

# ---------- 3. ä¸»çº¿ç¨‹å¾ªç¯è¯»è¾“å…¥ -------------
# ç­‰è¿æ¥çœŸæ­£å»ºç«‹å†è¿›å…¥äº¤äº’
while ws_app.sock is None or not ws_app.sock.connected:
    time.sleep(0.1)

try:
    while True:
        text = input("> ").strip()
        if text.lower() == "exit":
            ws_app.close()
            break
        # å…ˆæ¢è¡Œï¼Œé¿å…ä¸ä¸Šæ¡æµå¼å†…å®¹æ··åˆ
        print("\nğŸ’¬ AIï¼š", end="", flush=True)
        payload = [{"role": "user", "content": text}]
        ws_app.send(json.dumps(payload))
except KeyboardInterrupt:
    ws_app.close()
```
